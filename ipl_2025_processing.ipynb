IPL 2025 Data Pre Processing

Converting all JSON files to CSV files and making relationships between the data's.

```python
import pandas as pd
import json
```

1.Team Match Result Pre Processing

```python
with open('ipl_2025_json_files/ipl_2025_team_match_results.json') as f:
    data = json.load(f)

df_match = pd.DataFrame(data)
df_match.head()
```

```python
df_match.replace('Punjab Kings', 'PKBS', inplace = True)
df_match.head()
```

```python
num = 74
for index, row in df_match.iterrows():
    df_match.at[index, 'Scorecard'] = num
    num -= 1
df_match.rename({'Scorecard': 'Match Number'}, axis = 1, inplace = True)
df_match.head()
```

```python
df_match['Match Number'] = df_match['Match Number'].astype(int)
df_match.head()
```

```python
df_match.to_csv('ipl_2025_csv_files/match_results_ipl_2025.csv', index = False)
```
