IPL 2025 Data Pre Processing

Converting all JSON files to CSV files and making relationships between the data's.

```python
import pandas as pd
import json
import re
```

1.Team Match Result Pre Processing

```python
with open('ipl_2025_json_files/ipl_2025_team_match_results.json') as f:
    data = json.load(f)

df_match = pd.DataFrame(data)
df_match.head()
```

```python
df_match.replace('Punjab Kings', 'PKBS', inplace = True)
df_match.head()
```

```python
num = len(df_match)
for index,row in df_match.iterrows():
    df_match.at[index, 'Scorecard'] = num
    num -= 1
df_match.rename({'Scorecard': 'Match Number'}, axis = 1, inplace = True)
df_match.head()
```

```python
df_match['Match Number'] = df_match['Match Number'].astype(int)
df_match.head()
```

```python
df_match.to_csv('ipl_2025_csv_files/match_results_ipl_2025.csv', index = False)
```

2.Batting Summary Pre Processing

```python
with open('ipl_2025_json_files/ipl_2025_batting_summary.json') as f:
    data = json.load(f)

df_bat = pd.DataFrame(data)
df_bat.head()
df_bat.shape
```

```python
df_bat['dismissal'] = df_bat['dismissal'].fillna('')
df_bat['out/not out'] = df_bat['dismissal'].apply(lambda x: "out" if len(x) > 0 else "not out")
df_bat.head(20)
```

```python
df_bat['batsmanName'] = df_bat['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_bat['batsmanName'] = df_bat['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_bat.head(20)
```

```python
df_bat.drop(columns=['dismissal'],inplace=True)
df_bat.head()
```

```python
num = len(df_match)

for index, row in df_bat.iterrows():
    match_str = row['match']
    # Assign match numbers for playoffs
    if 'Final' in match_str:
        df_bat.at[index, 'Match Number'] = num
    elif 'Qualifier 2' in match_str:
        df_bat.at[index, 'Match Number'] = num - 1
    elif 'Eliminator' in match_str:
        df_bat.at[index, 'Match Number'] = num - 2
    elif 'Qualifier 1' in match_str:
        df_bat.at[index, 'Match Number'] =num - 3
    else:
        # For league matches like '70th Match', extract the number
        match_num = re.search(r'(\d+)(?=.. Match)', match_str)
        if match_num:
            df_bat.at[index, 'Match Number'] = int(match_num.group(1))
df_bat.head(1100)
```

```python
df_bat['Match Number'] = df_bat['Match Number'].astype(int)
df_bat.head(1100)
```

```python
df_bat.to_csv('ipl_2025_csv_files/batting_summary_ipl_2025.csv', index = False)
```
